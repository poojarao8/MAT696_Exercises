#### Advantages of inheritance:
    
1. Inheritance allows the child class to use features of the parent class while adding its own custom features.
2. It allows for code reuseability, and localizes the bugs and their fixes. 
3. Allows for better structring and readability (and hence sustainability) of your code.

Note: Lack of proper user and overuse can lead to a tangled mess of hierarchies.
